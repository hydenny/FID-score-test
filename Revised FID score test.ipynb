```python
from PIL import Image
import matplotlib.pyplot as plt
import imgaug.augmenters as iaa
from fid_score import *
```

```python
attribute = []

for i in range(5):
  attribute.append([])

for i in range(66,75):
  img = np.array(Image.open('imgs/%d-1.jpg' % i))

  img2 = np.array(Image.open('imgs/%d-2.jpg' % i))
  img3 = np.array(Image.open('imgs/%d-3.jpg' % i))
  img4 = np.array(Image.open('imgs/%d-4.jpg' % i))
  img5 = np.array(Image.open('imgs/%d-5.jpg' % i))
  img6 = np.array(Image.open('imgs/%d-6.jpg' % i))
  
  print('===========================================================')
  print(i)

  list = []  

  for j in range(2, 7):
    score = calculate_fid_given_paths(
    paths=[
        img,
        img %j
    ],
    batch_size=1,
    cuda=False,
    dims=2048
    )

    list.append(score)
    attribute[j - 2].append(score)

  for j in range(len(list)):
    print(list[j])

  print("Average: " + str(sum(list)/5))

#print(sum(attribute[2])/74)
#print(sum(attribute[3])/74)
#print(sum(attribute[4])/74)
#print(sum(attribute[5])/74)
#print(sum(attribute[6])/74)
```
